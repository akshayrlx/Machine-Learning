```python
import numpy as np   
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
import pandas as pd    
import matplotlib.pyplot as plt   
import seaborn as sns
```

```python
data=pd.read_csv('insurance.csv')
```

```python
data.head()
```

```python
data=data.replace('yes',1)
```

```python
data=data.replace('no',0)
```

```python
data
```

```python
sex=pd.get_dummies(data['sex'],drop_first=True)
region=pd.get_dummies(data['region'],drop_first=True)
```

```python
data=pd.concat([data,sex,region],axis=1)
```

```python
data
```

```python
data.drop(['sex','smoker','region'],axis=1,inplace=True)
```

```python
data
```

```python
data['male'] = data['male'].astype(int)
data['northwest'] = data['northwest'].astype(int)
data['southeast'] = data['southeast'].astype(int)
data['southwest'] = data['southwest'].astype(int)
```

```python
data
```

```python
x=data.drop(columns=['charges'],axis=1)
y=data['charges']
x=pd.DataFrame(x)
y=pd.DataFrame(y)
```

```python
x
```

```python
y
```

```python
from sklearn import preprocessing
X_scaled = preprocessing.scale(x)
X_scaled = pd.DataFrame(X_scaled, columns=x.columns)

y_scaled = preprocessing.scale(y)
y_scaled = pd.DataFrame(y_scaled, columns=y.columns)  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.30, random_state=1)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
ridge = Ridge(alpha=.3)
ridge.fit(X_train,y_train)
for i,col in enumerate(X_train.columns):   
    print ("Ridge model coefficients for {} is {}:".format(col,ridge.coef_[i]))
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
for i,col in enumerate(X_train):
    print ("Lasso model coefficients for {} is {}:".format(col,lasso.coef_[i]))
```

```python
print(regression_model.score(X_train, y_train))
print(regression_model.score(X_test, y_test))
```

```python
print(lasso.score(X_train, y_train))
print(lasso.score(X_test, y_test))
```
